```python
excludes = ['the', 'of', 'to', 'and', 'in', 'a', 'is', 'were', 'was', 'you',
            'I', 'he', 'his', 'there', 'those', 'she', 'her', 'their',
            'that', '[a]', '[b]', '[c]', '[d]', 'them', 'or','for','as',
            'are','on','it','be','with','by','have','from','not','they',
            'more','but','an','at','we','has','can','this','your','which','will',
            'one','should','points)','________','________.','all','than','what',
            'people','if','been','its','new','our','would','part','may','some','i',
            'who','answer','when','most','so','section','no','into','do','only',
            'each','other','following','had','such','much','out','--','up','these',
            'even','how','directions:','use','because','(10','time','(15','[d].',
            '-','it.','[b],','[a],','however,','1','c','1.','2.','b','d','a','(10',
            '2','12.','13.','29.','3.','4.','5.','6.','7.','8.','9.','10.','11.','14.',
            '15.']



def read_abstract(FILE):
    """给定路径，自动读取出所有摘要，合并成一个文本文件"""
    df = pd.read_csv(FILE, sep='\t')[['AU', 'TI', 'AB']]
    return ' '.join(
                    [ab['AB'] for _, ab in dd.iterrows() if isinstance(ab['AB'], str)] # 476 have abstract
                ) # str # 合并所有摘要成一个大字符串

def clean(text):
    """输入一个字符串，自动执行清洗文本过程，输出结果可以直接拿去进行词频统计"""
    # 转换小写
    text = text.lower()
    print("转换小写成功")
    # 去标点符号成功
    for ch in '!"@#$%^&*()+,-./:;<=>?@[]_`~{|}':
        text.replace(ch, ' ')
    print("去标点符号成功")
    # '...' 转换成 ['', '', ...]
    tokens = text.split(' ')
    # 去除停用词
    clean_tokens = tokens[:]
    stwords = stopwords.words('english')
    for token in tokens:
        if token in stwords:
            clean_tokens.remove(token)
            
    # 还原成词干
    porter_stemmer = PorterStemmer()
    clean_tokens = [porter_stemmer.stem(word) for word in clean_tokens]
    clean_tokens
    return clean_tokens

def count_(clean_text, verbose=True):
    """统计词频并排序"""
    # 词频统计
    counts = Counter(clean_text)
    # sort
    items = list(counts.items())
    items.sort(key=lambda x:x[1],reverse=True)
    for index, i in enumerate(range(3000)):
        word, count = items[i]
    #     print("{0:<10}{1:>5}".format(word,count))
        print(f"|{index:2}|\t{word:15}|  {count}") # 打印词频结果
    return items
    

def main():
    FILE = './data/acs_catal_2020.tsv'
    text = read_abstract(FILE)
    clean_text = clean(text)
    count_(clean_text)
```
